```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100) # 100 punkter mellan -5 till 5 
f = np. sin(x)

plt.figure(dpi = 100)
plt.plot(x, f)
plt.title("Sin(X)")
plt.xlabel("X")
plt.ylabel("X")
plt.show()
```

```python
#imperativ approche 

fig, ax = plt.figure(dpi=100), plt.axes()

ax.plot(x, f)
ax.set(xlim = (-6, 6), ylim= (-1.5, 1.5), xlabel = "x", ylabel = "f", title = "sin(x)")
```

```python
# Subplots

plt.figure(figsize= (16, 5), dpi= 100)
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.title("Plots")
plt.legend("sin(x), cos(x)")

plt.subplot(1, 2, 2)
plt.plot(x,np.cos(x))
plt.show()
```
