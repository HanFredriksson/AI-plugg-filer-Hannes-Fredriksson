### Anonymisera  data

Finns olika krypteringrar. 
- Dom som gåt att avkryptera.
- Doms som inte går att avkryptera.
- Två vägs eller en vägs kryptering.



```python
import hashlib as hl
# en sträng får samma hash o av sätt hur du kallar hash krypt på den
string_to_hash = "hanfre@live.se".encode() # skapar en binär sträng
string_to_hash_case = "Hanfre@live.se".encode()

hl.sha256(string_to_hash).hexdigest(), hl.sha256(string_to_hash_case).hexdigest() # hexdigest får vi en hexadecimal sträng

```

```python
import pandas as pd

df = pd.read_csv("../Data/BankChurners.csv", usecols=["CLIENTNUM", "Attrition_Flag", "Customer_Age", "Gender", "Total_Trans_Amt"])

```

```python
df.dtypes
```

```python
df["CLIENTNUM"] = df["CLIENTNUM"].astype(str)
hashs = df["CLIENTNUM"].apply(lambda x: hl.sha256(x.encode()).hexdigest())

hashs
```

```python
df.insert(1, "SHA Hash values", hashs)

df
```
