```python
import pandas as pd

data = dict(AI=25, Net=30, Java=27, App=56)

series = pd.Series(data=data)

print(series)
print(f"series[0]: {series[0]}")
print(f"series[1]: {series[1]}")
print(f"Series keys: {series.keys()}")
print(f"Series keys: {series[series.keys()[0]]}")

print(f"Series by colum: {series["AI"]}")
```

```python
import random as rng

rng.seed(42)

dice_series = pd.Series([rng.randint(1, 6) for _ in range(500000)])

print(dice_series)

print(f"Min Value {dice_series.min()}")
print(f"Max Value {dice_series.max()}")
print(f"Mean Value {dice_series.mean()}")
print(f"Median Value {dice_series.median()}")
print(f"Standard Value {dice_series.std()}")

```
