***Uppgift 5 Dice rolls convergence***

```python
import random as rnd
import matplotlib.pyplot as plt

#A
count_sixes = len([count for count in [rnd.randint(1, 6) for rolle in range(100)] if count == 6])
print(f"How many sixes in 100 roles: {count_sixes}")

#B
count_sixes_simulation = [len([count for count in [rnd.randint(1, 6) for rolle in range(10**rolled)] if count == 6])for rolled in range(1, 6)]
print(f"\nNumber of sixes rolled: {count_sixes_simulation}\n")

#C
probability = [sixes / 10**rolled for sixes, rolled in zip(count_sixes_simulation, range(1, 6))]
print (f"The probabilty of rolling a six: {probability}\n")

#C
num_rolls = (10, 100, 1000, 10000, 100000)
plt.plot(probability, '-*')
plt.title("Probability of six for different number of rolls")
plt.xticks([0,1,2,3,4], num_rolls);
plt.xlabel("Number of dice rolls")
plt.ylabel("Probability")
```

***Uppgift 6 Monto Carlo Simulation*** 

```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt 

def euclidean(x=1, y=2):
    distans = np.sqrt(np.square(x)+np.square(y))

    if distans < 1:
        return True
    else:
        return False

points_generator = [[rnd.uniform(-1.0, 1.0), rnd.uniform(-1.0, 1.0)] for simulations in range(5000)]

euclidean_sort_blue_x = [xy[0] for xy in points_generator if euclidean(*xy)]
euclidean_sort_blue_y = [xy[1] for xy in points_generator if euclidean(*xy)]

euclidean_sort_red_x = [xy[0] for xy in points_generator if not euclidean(*xy)]
euclidean_sort_red_y = [xy[1] for xy in points_generator if not euclidean(*xy)]

plt.scatter(euclidean_sort_red_x, euclidean_sort_red_y, c= "red")
plt.scatter(euclidean_sort_blue_x, euclidean_sort_blue_y, c= "blue")
plt.show()

```

***Uppgif 7 A cute rabbit among two ferocious snakes***

***#a***

Svar: Jag står med mitt val och öppnar dörren jag valt. För sannolikheten för att det är en orm bakom dörren är lika stor, oavsett om jag byter eller inte. 

```python
import random as rnd
import matplotlib.pyplot as plt

def sim_stay(n):

    picked = 0

    for i in range(n):
        
        doors = [1, 2, 3]
        pick = rnd.choice(doors)
        rabbit = rnd.choice(doors)

        if pick == rabbit:
            picked += 1
    
    return picked


def sim_switch(n):

    picked_switch = 0

    for i in range(n):

        doors = [1, 2, 3]
        pick = rnd.choice(doors)
        rabbit = rnd.choice(doors)

        for door in doors:
            if door != pick and door != rabbit:
                doors.remove(door)
                break
        
        
        pick = doors[0] if doors[0] != pick else doors[1]
        
        if pick == rabbit:
            picked_switch += 1
    
    return picked_switch

#stay = sim_stay(10)
switch = []
stay=[]
times_picked = (10, 100, 1000, 10000, 100000, 1000000)

for n in times_picked:
    stay.append(sim_stay(n)/n )
    switch.append(sim_switch(n)/n)

print(stay)
plt.plot(stay, '-*')
plt.plot(switch, '-*')
plt.title("Probability of six for different number of rolls")
plt.xticks([0,1,2,3,4,5], times_picked);
plt.xlabel("Number of dice rolls")
plt.ylabel("Probability")

```
