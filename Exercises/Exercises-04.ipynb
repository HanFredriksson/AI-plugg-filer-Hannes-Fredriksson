***Uppgift 5 Dice rolls convergence***

```python
import random as rnd
import matplotlib.pyplot as plt

#A
count_sixes = len([count for count in [rnd.randint(1, 6) for rolle in range(100)] if count == 6])
print(f"How many sixes in 100 roles: {count_sixes}")

#B
count_sixes_simulation = [len([count for count in [rnd.randint(1, 6) for rolle in range(10**rolled)] if count == 6])for rolled in range(1, 6)]
print(f"\nNumber of sixes rolled: {count_sixes_simulation}\n")

#C
probability = [sixes / 10**rolled for sixes, rolled in zip(count_sixes_simulation, range(1, 6))]
print (f"The probabilty of rolling a six: {probability}\n")

#C
num_rolls = (10, 100, 1000, 10000, 100000)
plt.plot(probability, '-*')
plt.title("Probability of six for different number of rolls")
plt.xticks([0,1,2,3,4], num_rolls);
plt.xlabel("Number of dice rolls")
plt.ylabel("Probability")
```
