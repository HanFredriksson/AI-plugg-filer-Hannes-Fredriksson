### Labb 3 VG uppgifter

```python
import Labb3_liner_classifier as clf
import matplotlib.pyplot as plt
import numpy as np

with open ("../Labs/Data/unlabelled_data.csv", "r") as unlable_data_r:
        data =[point.strip().split(",") for point in unlable_data_r.readlines()]
        data = [[float(point) for point in points] for points in data]

my_line = clf.LineClassifier(data)
lable_data = my_line.classifier()
```

```python
# Split the labeled data into two classes based on the label

class_1 = [(point[0], point[1]) for point in lable_data  if point[2] == 1]
class_0 = [(point[0], point[1]) for point in lable_data if point[2] == -1]
k_m = (-1.83, 0.234), (-0.489, 0),(-2, 0.16), (800, -120)

# Scatter plot for each class
plt.scatter(*zip(*class_1), label="Class -1", color='blue')
plt.scatter(*zip(*class_0), label="Class 1", color='purple')

# Gives the x value for plotting the decicion boundary
decicion_x = np.linspace(min(point[0] for point in lable_data), max(point[0] for point in lable_data), 100)

# y = kx + m decision  for the decicion boundary 
for k, m in k_m:
    decicion_y = k * decicion_x + m
    plt.plot(decicion_x, decicion_y, label=f"y = {k} * x + {m}")


plt.xlim(6, -6)
plt.ylim(6, -6)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Plot for decicion boundarys for VG")
```

```python
vg_line_1 = clf.LineClassifier(data, -0.489, 0)
vg_line_2 = clf.LineClassifier(data, -2, 0.16)
vg_line_3 = clf.LineClassifier(data, 800, -120)

vg_line_1.classifier()
plt.figure(figsize=(16,5), dpi=100)
plt.subplot(1, 2, 1)
vg_line_1.plot()

vg_line_2.classifier()
plt.figure(figsize=(16,5), dpi=100)
plt.subplot(1, 2, 2)
vg_line_2.plot()

vg_line_3.classifier()
plt.figure(figsize=(16,5), dpi=100)
plt.subplot(1, 2, 1)
vg_line_3.plot()
```

### Svar på VG frågorna

Som vi bara se på hur dom tre linjerna delar upp datan i märkena 0 och 1. Ser vi att i Linje 1 och 3 så faller en eller flera på punkter på linjen. I en linjär klassificerare vill man se så mycket separering mellan punkterna som möjligt. Det bästa hade varit skriva en kod som hittar den linje som har mest marginal mellan klusterna av punkter. Linje 2 två kommer nog närmast att vara den linje som separerar klusterna. För distansen mellan support vektorerna är som längst och linjen ser ut att ligga i mitten.
Det ger störst säkerhet att klassificera punkterna.
