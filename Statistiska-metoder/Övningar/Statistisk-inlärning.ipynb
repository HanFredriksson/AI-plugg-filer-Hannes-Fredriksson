```python
A = np.array(np.arange (16)).reshape ((4, 4))
print(A)
print(f"Seconde and fourth row:\n{A[[1 ,3]]}\n")
print(f"First and third columns:\n{A[:, [0 ,2]]}\n")
print(f"Try to get a sub-matrix from second and fourth row:\n{A[[1, 3], [0 ,2]]}\n")
print(f"A sub-matrix from second and fourth row:\n{A[[1, 3]][:, [0 ,2]]}\n")

# Easy way to get the sub-matrix, creating a meah objekt
# Here you can also use uneven number columns
idx = np.ix_([1, 3], [0, 2])
print(f"An easier way to get the sub-matrix:\n{A[idx]}\n")

# Third way is to use slices
# [Rows from : Rows to : steps, Elements from : Elemnets to : Steps]
print(f"Using slices:\n{A[1:4:2, 0:3:2]}\n")

```

```python
# Makes a vector from A first dimension and gives it a 0 repsentad by boolean
keep_rows = np.zeros(A.shape[0], bool)
print(f"{keep_rows}\n")

# Gives to elements bool = True
keep_rows[[1, 3]] = True
print(f"{keep_rows}\n") 

#Same as 0 = False or 1 = True
print(f"{np.all(keep_rows == np.array ([0,1,0,1]))}\n")

# Using Booleans we can retriev rows from an array
print(f"{A[keep_rows]}\n")

# We can alos creat a mesh to retrive rows and columns from an array
keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2 ,3]] = True
idx_bool = np.ix_(keep_rows, keep_cols)

print(f"{A[idx_bool]}\n")

# We can mix the mesh argumants with a list and n array of Booleans
idx_mixed = np.ix_([1, 3], keep_cols)
print(f"{A[idx_mixed]}\n")
```
