*Addition*

$P[A_1 \cup A_2] = P[A_1] + [A_2] - P[A_1 \cap A_2]$

Man tar bort snittet vid additon annars så räknar man snittet två gånger.

*Oberoende Händelser*

$A_1$ och $A_2$ är oberoende om endast och endast om villkoret 

$P[A_1 \cap A_2] = P[A_1] * P[A_2] =$ Om sannolikhet för snittet är snnolikheten fär händelserna multiplicerat med varandra är lika. Då är dom också oberoende.


*Beroende Händelser*

$P[A_2|A_1] =$ Givet att $A_1$ har hänt, vad är sannolikheten för $A_2$ = $\frac{P[A_1 \cap A_2]}{P[A_1]}$

Alt.

$P[A_2|A_1] = P[A_2] (P[A_1] \neq 0)$

$P[A_1|A_2] = P[A_1] (P[A_2] \neq 0)$


*Multiplikationsregeln*
Snittet för beroende händelser

$P[A_1 \cap A_2] = P[A_2|A_1] * P[A_1]$

$P[A_j|B] = \frac {P[B|A_j] * P[A_j]} {\sum_{i=1}^{N} P[B|A_j] * [A_j]}$
