## Vad är en Distribution?
En sannolikhetsdistribution beskriver hur sannolikheten är för ett utfall av en slumpmässig variabel.


*Kumulativ distributionsfunktionen*

Låt X vara en diskret s.v.

$F(x) = P[X \le x]$

F kallas för kumulativ distributionsfunktionen (CDF). Med andra ord: Vad är sannolikheten att X(Faktiska variabeln) antar värdet x(Förutspådda variabeln).

*Täthetsfunktion:*

$f(x) = P[X = x]$

Här tittar vi på en punkt i kurvan
1. $f(x) \ge 0$ för alla x
2. Suman av f(x) = 1 för alla x ($\sum_{x} f(x) = 1$)

*Uträkning av CDF*

$\sum_{x} F(x_0) = \sum_{x \le x_0} f(x)$



*Kontinuerliga*
Byter vi alla summa tecken till integral tecken och använder intergralkallkyl. 


## Väntevärde (Expectation)

*Average, mean, expected value*

*Genomsnitt, medel, väntvärde*

Viktigt bit i statstik, kallas en centvärde, ett genomsnitt. Gäller bara att sannolikhetes värdena är fullständiga. Väntvärde är ett av flera viktiga kvantitavia mått. De kallas moment (till en distribution). Kan säga att väntvärde är medelvärdet av villka värden X kan anta, viktas av sannolikheten för dess utfall .

## Varians (Variance)

Är också ett moment (till en distribution). Ger ett informellt mått på hur mycket en uppsättning (slump) tal är utspridda kring medelvärdet.
Låt X vara en s.v. med medel $\sigma$. Variansen av X, Var X, eller $\mu^2$

$Var X = \sigma^{2} = E[(x-\mu)^{2}] = E[x^2] - (E[x])^2$ ($\mu$ här är medelvärdet)

Har ingen enhet. För den kvadreras, så vi tappar enhteten och storleken tappas också. Detta vissar på spridningen på medelvärdet. Om vi har två varienser från två liknande data kan vi gör någon slutsats eller ge en tolkning av värdet. 

## Standardavvikelsen

Är bara roten ur Variansen. Nu kan vi igen genomför mellan olika tester och prov. Nu har vi enheter och storlek igen, för man har tagit roten ur Variansen. 

$\sigma = \sqrt{Var x} = \sqrt{\sigma^2}$

Standaravikelser är tillförlitilghete av Avvikelsen. I ML så sicktar vi på $2\sigma$. Ju mindre Standardavvikelsen är ju bättre är träffsäkerheten på X. 

Stickprovsavvikelsen(Standard error) i allmännafall ger 0 när datapunktern (utfall) går mot oändligheten. Denna Avvikelsen är mer applicerbar på data taget från något fysiskt. Medan Standaravvikelsen är med applicerbr på något abstrakt eller filsofiskt som en politisk åsikt eller åsikter.

## Gemensamma Distributoner (Joint Distributions)

Här finns det ett bettingat beroende mellan dom olika variablerna. Är viktigt inom ML för att kunna förutsäga stickprov.  
Här är det flera funktiner av flera variabler.

1. $f_{xy} (X,Y) \ge 0$
2. $\sum{x} \sum{y} f_{xy} (x, y) = 1$

*Marginel täthet*

Vi kan inte använda vanlig täthets funktioner. För att beräkna täthet måste vi fixer en av varibelerna. Vi fixerar ett x och summerar över alla möjliga y.  

$f_{x} (x) = \sum{y} f_{xy}(x,y)$

$f_{y} (y) = \sum{y} f_{xy}(x,y)$

*Bettingad täthets funnktion*

Här används dom marginella täthets funktionerna, $f_y(y)$. Vad den här gör är att hitta medlet för givet x. Om vi börjar plotta för alla x för en regressions kurva. Då har vi hitta medlet för alla förkomande x, i den gemensamma distributionen.  

$f_{x|y} (x) = \frac {f_{xy}(x,y)} {f_y(y)}$ (X)

$f_{y|x} (y) = \frac {f_{xy}(x,y)} {f_x(x)}$ (Y)

Givet ett X = x så räknar vi här medelt av Y för det x.et. 

Om vi plottar varje $x \in X$ mot Y får vi en regression av X på Y.

$\mu_{x|y} = \beta_{0} + \beta_{1}$ x är samma som kx+m ($\beta$ är våra parametrar här) Kommer vi hit kan vi rita en linje, en regressionslinje. Så länge detta är basserat på statestik, har bra variance och täcker större delen av populationen. Så kan vi förutsäga saker. 


En linjär regretion är en medelfördelning. Men kan bli riktigt fel i en Distibrution när punkterna är nära varandra. Leder till att vi måste veta i vilket intervall våran linjära regretion är gilltig. 

Låt $Y_{i}$ vara utfallen av en s.v. Y. 
Vikallar den responsvariabel. Forma en tabell.
$\beta$ kallas prediktorer

$Y_{i} = \beta_{0} + \beta_{1} x_{i} + E_{i}$

$E[E_{i}] = 0$

Detta ska tollkas vi gjort så massa mätningar av nånting Y och X. Där $E_i$ är en felterm med medlet = 0 

*Varje rad*

$y_{i} = \beta_{0} + \beta_{1} + \epsilon_{i}$

$\epsilon$ här är realliseringen. För $\beta_0$ och $\beta_1$ uppskattar vi värdet genom minsta kvadrat anpasningen.
$\epsilon_{i}$ "Hemlig", avvikelse från "sanna linjen", fel 

*Efter uppskattning*

b = är uppskattningen.

$y_{i} = b_{0} + b_{1} x_{i} + e_{i}$

Här har en faktisk felat.  
$e_{i}$ Kända avvikelser från regressinslinjen, en residual.
